```ruby
# phone number cleaning function
# input: string
# output: string
# 1. remove all non-numeric characters
# 2. if number is 10 digits, assume it's good
# 3. if number is 11 digits and the first number is 1, trim the 1 and use the first 10 digits
# 4. if the number is 11 digits and the first number is not 1, then it is a bad number
# 5. if the number is more than 11 digits, assume it is a bad number
# 6. if the number is less than 10 digits, assume it is a bad number
def clean_phone_num(number)
  number.gsub!(/\D/, '') # remove all non-numeric characters, and /\D/ means all non-numeric characters
  if number.length == 10
    '(' + number[0..2] + ')' + number[3..5] + '-' + number[6..9]
  elsif number.length == 11 and number[0] == '1'
    '(' + number[1..3] + ')' + number[4..6] + '-' + number[7..10]
  else number.length == 11 and number[0] != '1' or number.length < 10 or number.length > 11
    '(000)000-0000'
  end
end
 
```

```ruby
clean_phone_num('+1(518256.9886')
```

```ruby
def hour(dt_string)
  return DateTime.strptime(dt_string, '%m/%d/%y %H:%M').hour
end


require 'csv'
contents = CSV.open('../event_attendees_full.csv', headers: true, header_converters: :symbol)
contents.each do |row|
  pp hour(row[:regdate]).to_s + ':00    ' + row[:first_name] +  '_' + row[:last_name]
end
```

```ruby
def frequency_count(list)
  frequency = Hash.new(0)  # Create a new hash with default values of 0 to store the frequency count
  
  list.each do |item|  # Iterate over each item in the list
    frequency[item] += 1  # Increment the count for the current item in the frequency hash
  end
  
  return frequency  # Return the frequency hash containing the count for each item
end
```

```ruby
frequency_count([1, 1, 1, 2, 2, 2, 2, 2, 'a', 'C', 'a'])
```

```ruby
def count_frequency(array)
  array.max_by {|a| array.count(a)}
end
```

```ruby
count_frequency(['a', 'C', 'a'])
```
